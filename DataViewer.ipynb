```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import json
import cv2
import numpy as np
from PIL import Image
```

```python
def annotate_and_save_images(input_dir_path, output_dir_path, color_map):
    if not os.path.exists(output_dir_path):
        os.makedirs(output_dir_path)
    for root, _, files in os.walk(input_dir_path):
        for file in files:
            if file.lower().endswith(".jpg"):
                base_name = os.path.splitext(file)[0]
                img_path = os.path.join(root, file)
                json_path = os.path.join(root, base_name + ".json")
                
                # Read image using OpenCV
                img = cv2.imread(img_path)
                if img is None:
                    print(f"Error reading {img_path}. Skipping.")
                    continue
                
                # Original size (height, width)
                original_size = (img.shape[1], img.shape[0])
                new_size = (1015, 1015)
                
                # Resize the image using OpenCV
                img_resized = cv2.resize(img, new_size, interpolation=cv2.INTER_LANCZOS4)

                # Create a copy for drawing (we will draw directly on this image)
                annotated_img = img_resized.copy()

                # Load the JSON annotation file
                with open(json_path, 'r') as json_file:
                    annotations = json.load(json_file)
                    for shape in annotations['shapes']:
                        label = shape['label']
                        points = shape['points']
                        color = color_map.get(label, (0, 0, 255))  # Default color: red
                        
                        # Convert points to integer
                        points = np.array(points, dtype=np.int32)
                        
                        # Draw polygon based on points
                        cv2.polylines(annotated_img, [points], isClosed=True, color=color, thickness=2)
                        print(f"Drawn {label} at coordinates: {points[:2]}... ({len(points)} points)")

                # Save the annotated image as PNG
                output_img_path = os.path.join(output_dir_path, base_name + "_annotated.png")
                cv2.imwrite(output_img_path, annotated_img)

                print(f"Saved annotated image at {output_img_path}")


# Example usage:
input_dir_path_training  =r"/p/playground/Glioma-mdc2025/Data/OriginalData/Glioma_MDC_2025_training"
output_dir_path_training = r"/p/playground/Glioma-mdc2025/Data/annotated_data/training"

color_map = {
    "Mitosis":  (0, 0, 255),  # Red for not mitosis
    "Non-mitosis" : (0, 255, 0)   # Green for mitosis
}
annotate_and_save_images(input_dir_path_training, output_dir_path_training, color_map)





```

```python
input_dir_path_testing =r"/p/playground/Glioma-mdc2025/Data/OriginalData/Glioma_MDC_2025_test"
output_dir_path_testing = r"/p/playground/Glioma-mdc2025/Data/annotated_data/testing"
annotate_and_save_images(input_dir_path_testing, output_dir_path_testing, color_map)
```
